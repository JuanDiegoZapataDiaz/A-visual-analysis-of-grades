```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.linear_model import LinearRegression
import os
```

```python
pc1=pd.read_excel('pc1.xlsx')
```

# Score distribution: total

The total score distribution is unimodal, the median is close to 10 and we see that 50% of the scores are between 8 and 11. 

```python
fig, ax = plt.subplots(2, 1, figsize=(7, 5), gridspec_kw={"height_ratios": (2, 0.5)})
fig.suptitle('Notas PC1', fontsize=15)
sns.histplot(data=pc1, x='total', kde=True, ax=ax[0])
ax[0].set(xlabel=None)
sns.boxplot(data=pc1, x='total', ax=ax[1])
plt.show()
```

# Score distribution: Conceptual questions

```python
fig, ax = plt.subplots(2, 1, figsize=(7, 5), gridspec_kw={"height_ratios": (2, 0.5)})
fig.suptitle('Notas PC1: comentes (sobre 10 puntos)', fontsize=15)
sns.histplot(data=pc1, x='comentes', kde=True, ax=ax[0])
ax[0].set(xlabel=None)
sns.boxplot(data=pc1, x='comentes', ax=ax[1])
plt.show()
```
